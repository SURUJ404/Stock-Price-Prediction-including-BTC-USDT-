```python
import yfinance as yf
import datetime, quandl
from pandas_datareader import data, wb
```

```python
sp500 = yf.Ticker("^GSPC") ##be specific 
```

```python
sp500 = sp500.history(period="max")  
```

```python
sp500
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close",use_index=True)
```

```python
##Targetting the stock
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500  ## 1 for Up 0 for down
```

```python
# ##subsetting the datasets  
##Be Carefull
sp500 = sp500.loc["2010-01-01":].copy()  ## from the initial time to train
```

```python
sp500
```

```python
##Working with Random Forest for working with nonlinear data 
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

##Working with timeseries data
##cross validation dont use it leakage problem future to past : makes no sense
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close", "Volume" , "Open" , "High" , "Low"]
model.fit(train[predictors] , train["Target"])
```

```python
from sklearn.metrics import precision_score   ##Choose as per convenience matricews r or anything as per prediction accuracy

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)  ##precision score : currently 50% i.e 0.5
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

##Making it more robust
#Backtesting System

```python
##Last day data 

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model , predictors, start = 2500, step=250):
    all_predictions = []
    
    for i in range(start , data.shape[0] , step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train,test ,predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model , predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"]) ##above rows 3435+2552 
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
# last 2 days , last 5 weeks  , last 3 monmths trading day , __ so on 

new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column , trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .7] = 1  ##set the thresold according to yours 
    preds[preds < .7] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```
